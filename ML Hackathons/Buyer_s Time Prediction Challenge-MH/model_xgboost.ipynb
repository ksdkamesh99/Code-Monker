```
from xgboost import XGBRegressor
```

```
y_pred=model.predict(x_valid)
```

```
y_pred
```

```
y_pred=abs(y_pred)
```

```
import sklearn.metrics as m
```

```
m.mean_squared_log_error(y_valid,y_pred)
```

```
y_test=abs(y_test)
```
