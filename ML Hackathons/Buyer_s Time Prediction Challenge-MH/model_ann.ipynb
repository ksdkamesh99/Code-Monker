```
y_train=train['time_spent']
```

```
train.drop(['time_spent'],axis=1,inplace=True)
```

```
x_train=train
x_test=test
```

```
from xgboost import XGBRegressor
```

```
from sklearn.model_selection import train_test_split
```

```
x_trains,x_valid,y_trains,y_valid=train_test_split(x_train,y_train,random_state=7,test_size=0.1)
```

```
x_trains.shape
```

```
from keras.layers import Dense
from keras.models import Sequential
```

```
model=Sequential()
model.add(Dense(32,activation='relu',input_shape=[5]))
model.add(Dense(1,activation='linear'))
```

```
model.compile(optimizer='Adam',loss='mse')
```

```
model.summary()
```

```
model.fit(x_trains,y_trains,validation_data=(x_valid,y_valid),epochs=100,batch_size=8)
```

```
y_pred=model.predict(x_valid)
```

```
y_pred
```

```
y_pred=abs(y_pred)
```

```
import sklearn.metrics as m
```

```
m.mean_squared_log_error(y_valid,y_pred)
```
